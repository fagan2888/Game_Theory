```python
from egtplot import plot_animated, plot_static
```

```python
payoff_entriesWAR = [[1.0], [.8], [.6], [1.0], [.8], [.6], [.6], [.6], [.6]]
labels = ['A', 'W', 'R']
simplex = plot_static(payoff_entriesWAR, background=False, edge_eq=True, vert_labels=labels)
```

```python
payoff_entries_WAR_LoF = [[1.0], [.8], [.4], [1.0], [.8], [.5], [.6], [.6], [.6]]
labels = ['A', 'W', 'R']
simplex = plot_static(payoff_entries_WAR_LoF, background=False, edge_eq=True, vert_labels=labels)
```

```python
def get_payoff_WAR_GEN(h, l,omega,delta):
    return [[h, omega*h, delta*h], 
            [h, omega*h, delta*h], 
            [delta*h, delta*h, delta*h]]
```

```python
parameter_values = [[1], [.5], [.8], [.6]]
labels = ['A', 'W', 'R']
simplex = plot_static(parameter_values, custom_func=get_payoff_WAR_GEN, vert_labels=labels)
```

```python
simplex_WAR_GEN = plot_static(parameter_values, custom_func=get_payoff_WAR_GEN, vert_labels=labels,
                        paths=True)
```

```python
def get_payoff_WAR_HH_LoF(h, l,omega,delta,L_A,L_W):
    return [[h, omega*h, delta*h-L_A], 
            [h, omega*h, delta*h-L_W], 
            [delta*h, delta*h, delta*h]]
```

```python
parameter_values = [[1], [.5], [.8], [.6], [.2], [.1]]
labels = ['A', 'W', 'R']
simplex = plot_static(parameter_values, custom_func=get_payoff_WAR_HH_LoF, vert_labels=labels)
```

```python
simplex_WAR_GEN = plot_static(parameter_values, custom_func=get_payoff_WAR_HH_LoF , vert_labels=labels,
                        paths=True)
```

```python
simplex = plot_static(parameter_values, custom_func=get_payoff_WAR_HH_LoF, vert_labels=labels,
                        background=True)
```

```python
simplex = plot_static(parameter_values, custom_func=get_payoff_WAR_HH_LoF, vert_labels=labels,
                        paths=True, 
                        generations=10,
                        steps=2000,
                        ic_type='random',
                        path_color='viridis',
                        edge_eq=False)
```

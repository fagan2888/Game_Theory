```python
from egtplot import plot_animated
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

```python
def get_payoff_WAR_HH_LoF(h, l,omega,delta,L_A,L_W):
    return [[h, omega*h, delta*h-L_A], 
            [h, omega*h, delta*h-L_W], 
            [delta*h, delta*h, delta*h]]
```

```python
def get_payoff_WAR_HH(h, l,omega,delta):
    return [[h, omega*h, delta*h], 
            [h, omega*h, delta*h], 
            [delta*h, delta*h, delta*h]]
```

```python
parameter_values = [[1], [.5], [.8], [.6]]
animation, fps = plot_animated(parameter_values, custom_func=get_payoff_WAR_HH)
animation.ipython_display(fps=fps, loop=True, autoplay=True)
```

```python
parameter_values = [[1], [.5], [.8], [.6], [.2], [.1]]
labels = ['A', 'W', 'R']
animation, fps = plot_animated(parameter_values, custom_func=get_payoff_WAR_HH_LoF, vert_labels=labels)
animation.ipython_display(fps=fps, loop=True, autoplay=True)
```
